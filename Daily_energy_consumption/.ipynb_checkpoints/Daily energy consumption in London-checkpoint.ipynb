
The data provided is a time series from smart-meters in London from end of 2011 to 2014. It contains the daily consumption (kWh)
averaged over several households in London.

The last timestamp for which consumption is available is January 31st 2014.

It was built a forecasting model that can predict the consumption for February 2014(excluding 28th, so from the 1st to the 27th).
It was used prophet library, before was recognized seasons intervals and outliers like the winters and 2012 year. 

Usado prophet library, antes reconociendo intervalos estacionales, los inviernos en este caso y valores atípicos días por 2012


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from statsmodels.api import tsa
from sklearn.metrics import mean_absolute_error
from fbprophet import Prophet
import os
```

```python
# preprocess function: preprocess data and return ts: DataFrame to train and 
# ts_eval DataFrame to evaluate the prediction

def preprocess(df):

    # copy DataFrame
    df_copy = df.copy() 

    # Making column day like index of Datetime DataFrame
    # Asigno que la columna day sea el índice Datetime del dataframe
    df_copy['day'] = pd.DatetimeIndex(df_copy['day'])
    
    #checking the NaN in consumption colum"
    print("NaN in consumption colum",df_copy.consumption.apply(pd.isna).sum())
    #Like we have 27 NaN in consumption colum, it was replaced for the mean
    df_copy['consumption'] = df_copy['consumption'].fillna(df_copy['consumption'].mean())

    # I'll use Prophet library, so I need date column named ds and consumption named y
    #Prophet usa 2 columnas ds(fechas) y(dato q depende de fecha, a predecir), renombro las columnas
    df_copy = df_copy.rename(columns={'day': 'ds', 'consumption': 'y'}) 

    msk_eval = df_copy.evaluation_set
    df_copy.drop("evaluation_set", axis=1, inplace=True)
    
    # Split training/test data
    ts_eval = df_copy.loc[msk_eval]
    ts = df_copy[~msk_eval]
    
    
    return ts, ts_eval
   

```

```python
def train(ts):
    
    #seasonal effect: the winter, and the summer
    #Efecto estacional: winter: enero a abril, y summer, genero DataFrame con esos valores
     
    high_consumption_2012 = pd.date_range(start='2012-01-01', end='2012-04-01', freq='D')
    high_consumption_2013 = pd.date_range(start='2013-01-01', end='2013-04-01', freq='D')
    high_consumption_2014 = pd.date_range(start='2014-01-01', end='2014-04-01', freq='D')
  
     
    low_consumption_2012 = pd.date_range(start='2012-07-01', end='2012-09-01', freq='D')
    low_consumption_2013 = pd.date_range(start='2013-07-01', end='2013-09-01', freq='D')
    low_consumption_2014 = pd.date_range(start='2014-07-01', end='2014-09-01', freq='D')
    
    winter = pd.concat([pd.DataFrame(high_consumption_2012), 
    pd.DataFrame(high_consumption_2013), 
    pd.DataFrame(high_consumption_2014)],  
    ignore_index=True)
    
    summer = pd.concat([pd.DataFrame(low_consumption_2012), 
    pd.DataFrame(low_consumption_2013), 
    pd.DataFrame(low_consumption_2014)],  
    ignore_index=True)
    
    # Built Dataframe to Prophet
    # Adapto el DataFrame para ser usado por Prophet con columnas holiday y ds
    winters = pd.DataFrame({
    'holiday': 'Winter',
    'ds'     : winter[0]
    })
    summers = pd.DataFrame({
    'holiday': 'Summer',
    'ds'     : summer[0]
    })
    # outliers june 2012
    # rango atípicos junio 2012
    outlier_dates = pd.DataFrame(pd.date_range(start='2012-06-01', end='2012-07-01', freq='D'))
    
    outlier = pd.DataFrame({
    'holiday': 'Atypical',
    'ds'     : outlier_dates[0]
    })
    
    # concating outliers with winter
    # concateno DataFrames con rango de fechas estacionales con valores atípicos
   
    all_outlier_dates = pd.concat((winters, summers, outlier))
   
    #Prophet model
    forecast_model = Prophet(growth='linear',  weekly_seasonality=3, yearly_seasonality=3, holidays=all_outlier_dates)
    model = forecast_model.fit(ts)
    
    return model
```

```python
def predict(model):
    # I Building predict Dates DataFrame to 2014 february
    # adapto el rango de fechas en las que quiero predecir, generando un DataFrame del 01 al 27 de febrero 2014
    
    df_dates = pd.DataFrame(pd.date_range(start='2014-02-01', end='2014-02-27', freq='D'), columns=['ds'])
    
    # Predigo, y como prophet trae genera mucha columnas, la columna yhat es la predicción 
    preds = model.predict(df_dates)
    
    return preds
```

```python
#Loading dataset 

df = pd.read_csv("data.zip")
consumption= df.copy()
consumption.head(3)
```

```python
# Calling preprocess and train functions
ts, ts_eval = preprocess(consumption)
model = train(ts)
```

```python
# Calling predict function
model_predictions = predict(model)
model_predictions.tail(3)
```

```python
# graphic to look stationals seasonal intervals and outliers
# grafico para visualizar intervalos estacionales y valores atípicos
   
plt.figure(figsize=(12, 6))
plt.plot(ts.ds, ts["y"], "-o")
plt.xlabel('Date')
plt.ylabel('consumption')
```

The chart shows high consumption in winter and low consumption of energy in summer

# Interpreting and understanding the model

In the chart: the black dots correspond to the observed daily energy consumption, the blue chart represent the estimated  energy consumption in february 2014

# Comparison of mse without holidays 

```python
def train_without_holidays(df):             
    #Prophet model
    forecast_model = Prophet(growth='linear',  weekly_seasonality=3, yearly_seasonality=3)
    model = forecast_model.fit(df)
    return model

model_without_holidays = train_without_holidays(ts)
```

```python
model_predictions_without_holidays = predict(model_without_holidays)
```

```python
mean_squared_error(y_true=ts_eval.y, y_pred=model_predictions_without_holidays['yhat'])
```

We can see that the mean square error is higher without the holidays
